```python
df_nh = pd.read_excel("../data/sea-ice-coverage/Sea_Ice_Index_Daily_Extent_G02135_v3.0.xlsx", sheet_name=0)
df_nh.head()
```

```python
df_nh.rename(columns={"Unnamed: 0": "month", "Unnamed: 1": "day"}, inplace=True)
```

```python
df_nh.columns
```

```python
df_nh = df_nh.melt(id_vars="month", value_vars=df_nh.drop(["month", "day"], axis=1).columns, value_name="ice-extent", var_name="Year")
df_nh["hemisphere"] = "north"
```

```python
df_nh
```

```python
data_frame = pd.DataFrame({"month": [], "Year": [], "ice-extent": [], "hemisphere": []})
hemispheres = ["north", "south"]
for i in range(0, 2):
    df = pd.read_excel("../data/sea-ice-coverage/Sea_Ice_Index_Daily_Extent_G02135_v3.0.xlsx", sheet_name=i)
    df.rename(columns={"Unnamed: 0": "month", "Unnamed: 1": "day"}, inplace=True)
    df = df.melt(id_vars="month", value_vars=df.drop(["month", "day"], axis=1).columns, value_name="ice-extent", var_name="Year")
    df["Year"] = pd.to_numeric(pd.Series(df["Year"], dtype="Int64"), errors="coerce", downcast='signed')
    df = df.dropna().drop(columns=["month"], axis=1).groupby("Year").mean().reset_index().sort_values(by="Year", ascending=True)
    df["hemisphere"] = hemispheres[i]
    data_frame = pd.concat([data_frame, df],axis=0)
data_frame.drop(columns="month", inplace=True)
data_frame.to_csv("../data/sea-ice-coverage/seaice-treated.csv", index=False)
```
